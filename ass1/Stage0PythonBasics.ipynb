### What Are Loops?

Loops let you repeat an action multiple times without writing the same line again and again.

There are two main types in Python:

**for loops**
- Repeat a block for each item in a sequence
- Ideal for iterating over lists, strings, or ranges

**while loops**
- Repeat while a condition is true
- Useful when you don't know how many iterations you'll need in advance

Example â€” `for` Loop

````python
for i in range(5):
    print("Iteration:", i)
````

Understanding `range()`

You can think of `range(5)` as generating numbers `[0, 1, 2, 3, 4]`.

**Key points about `range()`:**
- Starts counting from 0 by default
- Stops before reaching the specified number
- Creates a sequence of integers

**Examples:**
```python
range(5)    # â†’ 0, 1, 2, 3, 4
range(3)    # â†’ 0, 1, 2
range(1, 5) # â†’ 1, 2, 3, 4

Example â€” Print Even Numbers

````python
for i in range(1, 11):
    if i % 2 == 0:
        print(i, "is even")
````

Example â€” while Loop

A while loop continues running as long as a condition is true.

**How while loops work**:
- Check condition before each iteration
- Execute code block if condition is `True`
- Repeat until condition becomes `False`

````python
count = 5
while count > 0:
    print("Countdown:", count)
    count -= 1  # Decrease by 1 each time
print("Blast off! ðŸš€")
````

**Understanding `count -= 1`**

The expression `count -= 1` is a shorthand assignment operator that decreases the value of `count` by 1.

What it means:
- **Long form**: `count = count - 1`
- **Short form**: `count -= 1`

Example â€” Compute Factorial of a Number

The factorial of a number n is the product of all positive integers up to n.
For example:

5!=5Ã—4Ã—3Ã—2Ã—1=120

Letâ€™s implement it using a function:

````python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print("5! =", factorial(5))
````

### Next, Iâ€™ll modify the factorial function
### This one will handle 0 and negative numbers properly

````python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result
````

### Iâ€™ll test the factorial function with a few examples

````python
print("Factorial of -10:", factorial(-10))
print("Factorial of 0:", factorial(0))
print("Factorial of 6:", factorial(6))
````

### Finally, Iâ€™ll write the Fibonacci function
### It returns a list of the first n Fibonacci numbers

````python
def fibonacci(n):
    if n <= 0:
        return []  # no Fibonacci numbers for non-positive input
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        seq = [0, 1]
        while len(seq) < n:
            seq.append(seq[-1] + seq[-2])
        return seq
````

### Testing the Fibonacci function

````python
print("First 10 Fibonacci numbers:", fibonacci(15))
````

## Input/Output & File Basics

### What Is Input/Output (I/O)?

Input/Output (I/O) is how your program receives data and sends data out.

- Input: Getting information from the user or a file.

- Output: Displaying results on the screen or saving them to a file.

Using `input()` for User Input

The `input()` function lets you ask the user for a value. By default, it returns a string.

### Writing to a File

You can save text to a file using the `open()` function with mode 'w' (write):

````python
# Open a file for writing
file = open("example.txt", "w")
file.write("Hello, Python!\n")
file.write("This is a test.\n")
file.close()
````

- 'w' mode overwrites the file if it already exists.
- The second argument is the mode:

     "r" â†’ read (default)

     "w" â†’ write (overwrite)

     "a" â†’ append (add to the end)

- Always close the file to save changes.

### Reading from a File

````python
# Open a file for reading
file = open("example.txt", "r")
content = file.read()
print(content)
file.close()
````

You can also read line by line:

````python
file = open("example.txt", "r")
for line in file:
    print("Line:", line.strip())
file.close()
````

**Tip**: `.strip()` removes extra spaces or newline characters.

**Using with Statement (Best Practice)**

Python provides with to automatically handle closing files:

````python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
````
